### Import Libraries

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time
```

### Initialize web driver and data frames

```python
s = Service('chromedriver.exe')
driver = webdriver.Chrome(service = s)

driver.get('https://ph.indeed.com/')
time.sleep(2)

search_box = driver.find_element(By.XPATH, '//*[@id="text-input-what"]')
search_box.send_keys('data engineer')
time.sleep(2)

search_box.send_keys(Keys.ENTER)
time.sleep(2)

job_headers = ['Job Title', 'Company', 'Company Rating', 'Location', 'Salary', 'Job Description']
job_table = pd.DataFrame(columns = job_headers)

company_headers = ['Company', 'Industry', 'CEO', 'Company Size', 'Revenue']
company_table = pd.DataFrame(columns = company_headers)
```

### Web crawler

```python
while True:
    
    postings = driver.find_elements(By.CLASS_NAME, 'job_seen_beacon')
    
    for post in postings:
        
        try:
            post.click()
            time.sleep(5)
        except:
            pass

        #Get HTML code of website
        soup = BeautifulSoup(driver.page_source, 'lxml')

        #Job Title
        title = soup.find('h1', class_ = 'icl-u-xs-mb--xs icl-u-xs-mt--none jobsearch-JobInfoHeader-title is-embedded').text
        title = (' ').join(title.split(' ')[:-3])
        time.sleep(1)

        #Company Name
        try:
            xpath = '//*[@id="jobsearch-ViewjobPaneWrapper"]/div/div/div/div/div/div/div[1]/div/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[2]/div/a'
            company = driver.find_element(By.XPATH, xpath).text
        except:
            try:
                xpath = '//*[@id="jobsearch-ViewjobPaneWrapper"]/div/div/div/div/div/div/div[1]/div/div[1]/div[1]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/a'
                company = driver.find_element(By.XPATH, xpath).text
            except:
                xpath = '//*[@id="jobsearch-ViewjobPaneWrapper"]/div/div/div/div/div/div/div[1]/div/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[2]/div'
                company = driver.find_element(By.XPATH, xpath).text
        time.sleep(1)

        #company rating
        try:
            company_rating = soup.find('a', class_ = 'icl-Ratings-starsCountWrapper icl-Ratings-link').get('aria-label')
        except:
            company_rating = ''
        time.sleep(1)

        #Location
        try:
            xpath = '//*[@id="jobsearch-ViewjobPaneWrapper"]/div/div/div/div/div/div/div[1]/div/div[1]/div[1]/div[1]/div[2]/div/div/div/div[2]/div'
            location = driver.find_element(By.XPATH, xpath).text
            if '•' in location:
                location = ' '.join(location.split('•')[:-1])
        except:
            xpath = '//*[@id="jobsearch-ViewjobPaneWrapper"]/div/div/div/div/div/div/div[1]/div/div[1]/div[1]/div[2]/div[2]/div/div/div/div[2]/div'
            location = driver.find_element(By.XPATH, xpath).text
            if '•' in location:
                location = ' '.join(location.split('•')[:-1])
        time.sleep(1)
        
        #Salary
        try:
            xpath = '//*[@id="salaryInfoAndJobType"]/span'
            salary = driver.find_element(By.XPATH, xpath).text
        except:
            salary = ''
        time.sleep(1)
        
        #Job Description
        xpath = '//*[@id="jobDescriptionText"]'
        job_description = driver.find_element(By.XPATH, xpath).text
        time.sleep(1)
        
        #Add to job_table dataframe
        row = pd.DataFrame({'Job Title':[title], 'Company':[company], 'Company Rating':[company_rating], 'Location':[location], 'Salary':[salary], 'Job Description':[job_description]})
        job_table = pd.concat([job_table, row], ignore_index=True)
        time.sleep(1)
        
        #Go to company profile (new window)
        try:
            try:
                xpath = '//*[@id="jobsearch-ViewjobPaneWrapper"]/div/div/div/div/div/div/div[1]/div/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[2]/div/a'
                driver.find_element(By.XPATH, xpath).click()

            except:
                xpath = '//*[@id="jobsearch-ViewjobPaneWrapper"]/div/div/div/div/div/div/div[1]/div/div[1]/div[1]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/a'
                driver.find_element(By.XPATH, xpath).click()
            time.sleep(3)

            #Switch to new window
            driver.switch_to.window(driver.window_handles[1])
            time.sleep(2)

            company_page = BeautifulSoup(driver.page_source, 'lxml')

            #Company Profile

            #CEO
            try:
                ceo = company_page.find('span', class_ = 'css-1w0iwyp e1wnkr790').text
            except:
                ceo = ""
            time.sleep(1)

            #company size
            try:
                company_size = company_page.find_all('div', class_ = 'css-1k40ovh e1wnkr790')[0].text
            except:
                company_size = ""
            time.sleep(1)

            #revenue
            try:
                revenue = company_page.find_all('div', class_ = 'css-1k40ovh e1wnkr790')[1].text
            except:
                revenue = ""   
            time.sleep(1)

            #Industry
            try:
                if len(company_page.find_all('div', class_ = 'css-1w0iwyp e1wnkr790')) == 1:
                    industry = company_page.find_all('div', class_ = 'css-1w0iwyp e1wnkr790')[0].text
                else:
                    industry = company_page.find_all('div', class_ = 'css-1w0iwyp e1wnkr790')[1].text
            except:
                industry = ""
            time.sleep(1)

            #foundation date
            try:
                if len(company_page.find_all('div', class_ = 'css-1w0iwyp e1wnkr790')) == 2:
                    foundation_date = company_page.find_all('div', class_ = 'css-1w0iwyp e1wnkr790')[0].text
            except:
                foundation_date = ""

            #Add to company dataframe
            row = pd.DataFrame({'Company':[company], 'Industry':[industry], 'CEO':[ceo], 'Company Size':[company_size], 'Revenue':[revenue]})
            company_table = pd.concat([company_table, row], ignore_index=True)
            time.sleep(1)

            #close new window
            driver.close()
            time.sleep(1)

            #return to original window
            driver.switch_to.window(driver.window_handles[0])
            time.sleep(3)
       
        except:
            pass

    try:
        driver.find_element(By.CSS_SELECTOR, '[aria-label="Next Page"]').click()
        time.sleep(5)
    except:
        break
```
